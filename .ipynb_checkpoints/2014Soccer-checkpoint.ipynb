```python
import json
import requests
import random
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
soccer_data_df = pd.read_csv("data/team_data.csv")
soccer_data_df.head()
```

```python
plt.scatter(soccer_data_df["WikiStanding"], soccer_data_df["Short Passes per Game"], c="blue", alpha=0.7, linewidth=1, marker="o", edgecolors="black")
plt.grid(True)
plt.xlim(35, 0)
plt.xlabel("Ranking")
plt.ylabel("Miles per Game")
plt.title("Ranking vs. Miles per Game")
plt.show()
                        
    
```

```python
plt.scatter(soccer_data_df["WikiStanding"], soccer_data_df["Miles per Game"], c="blue", alpha=0.7, linewidth=1, marker="o", edgecolors="black")
plt.grid(True)
plt.xlim(35, 0)
plt.xlabel("Ranking")
plt.ylabel("Miles per Game")
plt.title("Ranking vs. Miles per Game")
plt.show()
                        
    
```

```python
plt.figure(figsize=(20,10))

shots_per_game = soccer_data_df["Shots per Game"]
x_axis = np.arange(len(shots_per_game))
x_axis

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, soccer_data_df["Team"], rotation=70)

# Sets the x limits of the current chart
#plt.xlim(-0.75, len(x_axis)-0.25)

# Sets the y limits of the current chart
#plt.ylim(0, max(users)+5000)

# Give our chart some labels and a tile
plt.title("Team Shots per Game")
plt.xlabel("Team")
plt.ylabel("Shots per Game")

plt.bar(x_axis, shots_per_game, color='r', alpha=0.5, align="center")

# Print our chart to the screen
plt.show()
```

```python
#team_shots = soccer_data_df.loc[(soccer_data_df["Advanced"] == "Y")]
team_shots = soccer_data_df[["Team","Shots per Game", "Advanced"]]

team_shots = team_shots.set_index("Team").sort_values("Shots per Game", ascending=False)
   
    
team_shots.plot(kind="bar", figsize=(20,10), rot=70, width=.7)

plt.title("Team Shots per Game")

plt.show()
```

```python
# Labels for the sections of our pie chart
labels = ["AFC","CAF","CONCACAF","CONMEBOL","OFC","UEFA"]

# The values of each section of the pie chart
sizes = [1, 0, 1, 22, 0, 56]

# The colors of each section of the pie chart
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue", "yellow", "green", "blue"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0, 0, 0, 0.1)

# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
# enables auto percent you to display percent values using python formatting 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")
plt.title("Top 4")

# Prints our pie chart to the screen
plt.show()
```
